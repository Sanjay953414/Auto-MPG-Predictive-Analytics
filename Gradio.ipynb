**Importing Important Libraries**

```python
import pickle
import gradio as gr
import pandas as pd
```

**Loading Saved Model**

```python
#Loading Saved Model using Pickle
with open("Fuel_Pred1.0.pkl","rb") as f:
    model = pickle.load(f)
```

***Creating Prediction Function***

```python
# Define the prediction function
def predict_mpg(cylinders,displacement,horsepower, weight,acceleration,model_year,origin):
    # Create a dataframe with the input values
    input_data = pd.DataFrame({'cylinders': [cylinders], 'displacement': [displacement], 'horsepower': [horsepower], 'weight': [weight], 'acceleration': [acceleration]})
    
    # Use the model to make a prediction
    prediction = model.predict(input_data)
    
    # Return the predicted rainfall
    return f"The mpg is {prediction[0]:.2f}"

# Create the Gradio interface
# Create the Gradio interface
iface = gr.Interface(
    fn=predict_mpg,
    inputs=[gr.Number(label="cylinders", precision=0, value=8, minimum=1, maximum=12),gr.Number(label="displacement", precision=0, value=307, minimum=1, maximum=1000),gr.Number(label="horsepower", precision=0, value=130, minimum=1, maximum=1000),gr.Number(label="weight", precision=0, value=3504, minimum=1, maximum=10000),gr.Number(label="acceleration", precision=0, value=12, minimum=1, maximum=100)],outputs=gr.Textbox(label="prediction"))

# Launch the Gradio interface
iface.launch()
```
